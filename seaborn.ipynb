```python
 import seaborn as sns
 	# Apply the default theme
sns.set_theme()
 	# Load an example dataset
tips=sns.load_dataset("tips")
tips.info()
tips.head(5)
tips.tail(5)
 	# Create a visualization
sns.scatterplot(x="total_bill", y="tip", data=tips) 
```
